# Load dataset and imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("outputs/datasets/collection/Data-Melbourns_F_fixed.csv", parse_dates=["date"])

# Scatter plot: Average Temperature vs COD
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Average Temperature", y="Chemical Oxygen Demand", data=df)
plt.title("COD vs Average Temperature")
plt.xlabel("Average Temperature (°C)")
plt.ylabel("Chemical Oxygen Demand (mg/L)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Correlation calculation
correlation = df["Average Temperature"].corr(df["Chemical Oxygen Demand"])
print(f"Pearson correlation between Average Temperature and COD: {correlation:.2f}")


import streamlit as st

st.subheader("Hypothesis 1: Temperature vs COD")
st.markdown(f"""
**Correlation between Average Temperature and COD:** {correlation:.2f}

**Conclusion:**  
The correlation between temperature and COD is very weak (**{correlation:.2f}**), suggesting no meaningful relationship between higher temperatures and lower COD levels.  
This does not support the original hypothesis that COD decreases in warmer conditions.
""")
```

---

## Summary of Hypothesis Test

**Tested Hypothesis:**  
Higher temperatures are associated with lower Chemical Oxygen Demand (COD) levels due to increased microbial activity.

**Conclusion:** We **fail to reject the null hypothesis**. There is **no statistically significant** correlation between temperature and COD levels. The observed relationship is weak, with no consistent downward trend in COD as temperature increases.

**Business Insight:**  
- While temperature was hypothesised to reduce COD via microbial activity, the data does not support a strong or actionable seasonal pattern.  
- This result helped narrow the focus of feature selection in later modelling stages, allowing prioritisation of other operational variables.
